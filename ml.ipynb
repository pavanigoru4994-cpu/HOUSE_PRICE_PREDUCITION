#### Data preprocessing

data.isnull().sum(): It is used to count the missing values or NaN values in every column 

```python
data.isnull().sum()
```

 * first we take the Washrooms column from the dataset
* pd.to_numeric(data['Washrooms']): it is used to convert the every value into numeric type
* errors='coerece': it is used to when we have invalid values like string format values then it will be converted into NaN Values
data
* ['Washrooms']= : The converted numerical values are stored in the new column called Washrooms 

```python
data['Washrooms']=pd.to_numeric(data['Washrooms'],errors='coerce')
```

* data['Washrooms']: we take the washrooms column
* fillna(data['Washrooms']): The NaN values are replaced with the median of the numeric value
* median(): it returns the middle value of the numeric column
* data['Washrooms']= : the column with replaced missing values is saved in the original column 

```python
data['Washrooms']=data['Washrooms'].fillna(data['Washrooms'].median())
```

* Here we use again data.isnull.sum() is to check again washrooms data  before we have null values. Now we are again checking washrooms data will be changed into null values. 

```python
data.isnull().sum()
```

```python
data['Tennants']=pd.to_numeric(data['Tennants'],errors='coerce')
```

```python
data['Tennants']=data['Tennants'].fillna(data['Tennants'].median())
```

```python
data.isnull().sum()
```

```python
data['Area']=data['Area'].astype(str).str.extract('(\d+)')
```

```python
data['Area']=pd.to_numeric(data['Area'],errors='coerce')
```

```python
data['Area']=data['Area'].fillna(data['Area'].median())
```

```python
data[['Washrooms','Area']].isnull().sum()
```

```python
data['Price'] = (
    data['Price']
    .astype(str)
    .str.replace(r'[^\d]','',regex=True)
)
```

```python
data['Price']=data['Price'].astype(float)
```

```python
data['Price'].dtype
```

```python
data['Price'].apply(type).value_counts()
```

```python
x = data.drop('Price', axis=1)
y = data['Price']
```

```python
x = pd.get_dummies(x,drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

### Splitting the data

```python
from sklearn.model_selection import train_test_split
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print(x_train.dtype)
print(y_train.dtype)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)   

print("MAE :", mae)
print("MSE :", mse)
print("RMSE :", rmse)
print("R2 Score :", r2)
```

```python
new_house = pd.DataFrame({
    'Area' : [1400],
    'Bedrooms' : [6],
    'Washrooms' : [3]
})

new_house = pd.get_dummies(new_house)
new_house = new_house.reindex(columns=x.columns, fill_value=0)

new_scaled = scaler.transform(new_house)

predicted_price = model.predict(new_scaled)
print("predicted House Price:",predicted_price[0])
```
